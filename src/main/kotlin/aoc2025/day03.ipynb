```kotlin
val testInput = """
987654321111111
811111111111119
234234234234278
818181911112111""".trim()
```

```kotlin
fun String.toNumbers(): List<List<Int>> = lines().map {
    it.toCharArray().map { it.digitToInt() }
}
val testNumbers: List<List<Int>> = testInput.toNumbers()
val numbers = today.input().toNumbers()
```

```kotlin
fun findMaxAndIndex(list: List<Int>): Pair<Int, Int> {
    var max = 0
    var pos = 0

    list.forEachIndexed { index, i ->
        if (max < i) {
            max = i
            pos = index
        }
    }
    return max to pos
}

fun solvePartOne(input: List<List<Int>>): Int {
    return input.map { line ->

        val (left, leftIndex) = findMaxAndIndex(line.subList(0, line.size - 1))
        val right = line.subList(leftIndex + 1, line.size).max()
        "$left$right".toInt()
    }.sum()
}
```

```kotlin
println(solvePartOne(testNumbers))
println(solvePartOne(numbers))
```

```kotlin
fun solvePartTwo(input: List<List<Int>>, batterySize: Int): Long {
    return input.map { line ->

        var joltageString = ""
        var firstIndex = 0
        (batterySize - 1 downTo 0).map { i ->
            val (left, leftIndex) = findMaxAndIndex(line.subList(firstIndex, line.size - i))
            firstIndex += leftIndex + 1
            joltageString += left
        }
        joltageString.toLong()
    }.sum()
}
```

```kotlin
println(solvePartTwo(testNumbers, batterySize = 12))
println(solvePartTwo(numbers, batterySize = 12))
```
